```python
def count(df):
    cnt =0
    for i in df:
        if i ==0:
            cnt = cnt+1
    return cnt
```

K-NN, random forest and Naiive Bayes can deal with missing values.  the k-NN algorithm considers the missing values by taking the majority of the K nearest values. 

Below is a list of the applicable algorithms with explanation:


Mean/median imputation: In this method, missing values are replaced by the mean or median value of the available data. This method is simple and easy to implement but can lead to biased estimates if the data has a significant number of missing values or if the missing values are not missing at random.

K-nearest neighbor imputation: In this method, missing values are replaced with the average of the K closest neighboring data points. This method can be effective if the data has a complex structure and the missing values are not clustered together.

Multiple imputation: In this method, missing values are imputed multiple times using a statistical model, and the results are combined to estimate the final values. This method can provide more accurate estimates than other methods but requires more computational resources.

Maximum likelihood estimation: This method involves estimating the missing values based on the likelihood function of the observed data. This method can be effective if the data follows a specific probability distribution.

Expectation-maximization algorithm: This method is a variant of maximum likelihood estimation and is used to estimate the missing values in a dataset with a complex structure.
