Box-Cox transform approach

```python
from scipy import stats
import seaborn as sns

rp = df['resting blood pressure']
fitted_data, fitted_lambda = stats.boxcox(rp)
fig, ax = plt.subplots(1, 2)
sns.distplot(rp, hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 2},label = "Non-Normal", color ="green", ax = ax[0])

sns.distplot(fitted_data, hist = False, kde = True,kde_kws = {'shade': True, 'linewidth': 2},label = "Normal", color ="green", ax = ax[1])
plt.legend(loc = "upper right")


fig.set_figheight(5)
fig.set_figwidth(10)

print(f"Lambda value used for Transformation: {fitted_lambda}")
```

```python
(pd.DataFrame(fitted_data)).skew(axis = 0, skipna = True)
```

box cox did best, then logarithm, and then square root.

Log transform: The log transform is used to reduce the effect of extreme values and to make the data more normally distributed. Since it takes the log of each value, it compresses the range of values at the high end and expands the range at the low end. The log transform is particularly useful when dealing with data that has a skewed distribution, such as income or population data.

Square root transform: The square root transform is similar to the log transform in that it can be used to reduce the effect of extreme values and make the data more normally distributed. However, the square root transform is less powerful than the log transform in this regard. It is often used when the data has a positive skew and the values are all positive.

Box-Cox transform: The Box-Cox transform is a family of power transformations that includes both the log and square root transformations as special cases. It is used to find the best power transformation for a given dataset. The Box-Cox transform can handle both positive and negative data, and it can be used to transform data that has a wide range of distributions.

In summary, the log transform is the most powerful transformation for reducing the effect of extreme values and making data more normally distributed. The square root transform is less powerful but still useful, and it is often used when the data has a positive skew and all values are positive. The Box-Cox transform is a more flexible approach that can handle a wider range of distributions and can be used to find the best transformation for a given dataset.
