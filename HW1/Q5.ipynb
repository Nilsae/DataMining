```python
import numpy as np 
import seaborn as sns
%matplotlib inline
df = pd.read_csv('data/Smartphone.csv')
correlations = df.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlations,square=True,ax = ax)
```

If we sort the pairs of attributes with high correlation, 
ram and price range have the highest correlation;
then fc(Front Camera mega pixels) and pc(Primary Camera mega pixels);
and then three_g(has 3G or not) and four_g(has 4G or not);

Then, px_width and px_height; and screen width and screen height seem to have some(but not large) correlation. 

There are also very small correlations beween battery power and price range, 
between px height and price range, 
and between px width and proce range. 
 
Not much correlation is detected between other attributes.
